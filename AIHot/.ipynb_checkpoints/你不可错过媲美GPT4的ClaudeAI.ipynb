## 一、Anthropic简介：



首先，Claude 是 [Anthropic](https://www.anthropic.com/) 公司的一款 AIGC 产品。

Anthropic 对应一个定理 —— 人则定理，那什么是人则定理？

> 人择原理，是一种认为【物质宇宙必须与观测到它的存在意识的智慧生命相匹配】的哲学理论。
> 
> 
> 有些支持者提出人择原理解释了宇宙的年龄和为什么物理常数能够保证有意识生命的存活。
> 
> 他们也认为这个宇宙能给予智能生命（可观测者）存活的那么高的标准是一件正常的事情。 —— 维基百科
> 

白话上理解，**大概就是：并不是宇宙创造了人类，而是人类和宇宙互相选择。**

这个过程就像现在的机器人一样，不全是人创造了机器人，也是机器人选择了适应人类。


## 二、Claude介绍

Claude 有什么不一样？

**一句话：Claude 强于 GPT-3，逊于 GPT-4；**

最要的是：**它借助 slack 这个成熟平台，无需审核，目前还是免费、不限次数！在网页端、手机端都能直接使用。**

它被作为一款应用内嵌到其它程序里面，这样就可以直接应用，而不是去到一个特定的网页呼起 AIGC 聊天会话。

## 三、AI服务

- Claude提供了两种访问方式，API和Slack机器人
- API访问方式需要申请并等待通过，通过后测试时的并发限制为1，用于生产时需要与Anthropic协商具体限速
- Claude目前有两个版本，一个是**功能强大的版本Claude**，擅长于从复杂的对话和创造性的内容生成到详细的指令跟随的广泛任务；一个是**速度更快，价格更优的Claude Instant**，它也可以处理随意对话、文本分析、摘要和文档问答
- 两个版本的Claude prompt与output分开计费，能够处理内容的最长长度、收费标准参考官网

## 四、特点：

- 1、支持长达10万字符的上下文（超长）
- 2、对比ChatGPT，Claude的响应速度更快
- 3、与ChatGPT类似，**Claude能够处理多语言、不能联网**；
- 4、支持个性化finetune，但需要与Authropic联系Claude
- 5、目前不支持embedding，Anthropic推荐使用SBERT生成句子embedding
- 6、Claude in Slack目前是beta版本

Claude目前存在一些缺陷，包括以下几点：

- 1、Claude经常在复杂的算术和推理中出错
- 2、Claude有时会产生幻觉或编造信息和细节
- 3、Claude没有联网
- 4、Claude的训练数据是两年前的
