## 2、前向传播
1. 先计算$h_1$的所有输入：$net_{h_i}=w_1 * i_1 + w_2*i_2+b_1*1$，代入数据可得：;
    
    $net_{h_1}=0.15∗0.05+0.2∗0.1+0.35∗1=0.3775$
    
    
2. 然后利用logistic函数计算得$h_1$的输出:
    
    $out_{h_1}= \frac{1}{1+e^{-net_{h_1}}}=\frac{1}{1+e^{-0.3775}}=0.593269992$
    

3.  用同样的方法得:

    $out_{h_2}$=0.596884378
    
对输出层神经元重复这个过程，使用隐藏层神经元的输出作为输入。这样就能给出o1的输出：

$net_{o_1}=w_5*out_{h_1}+w_6*out_{h_2}+b_2*1$，代入数据可得：

neto1=0.4∗0.593269992+0.45∗0.596884378+0.6=1.105905967，则其输出为：

$out_{o_1}= \frac{1}{1+e^{-net_{h_1}}}=\frac{1}{1+e^{-0.1.105905967}}=0.75136507$

同样可以得到$out_{o_2} = 0.772928465$

### 2.1、开始统计所有的误差

如上图，o1的原始输出为0.01，而神经网络的输出为0.75136507，则其误差为：
$E_{o_1}=\sum \frac{1}{2}(0.01-0.75136507)^2=0.298371109$

同样可得:$E_{o_2}=0.023560026$

综合所述，可以得到总误差为：$E_total=E_{o_1}+E_{o_2}=0.298371109$
